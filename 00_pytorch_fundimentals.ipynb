## Reshaping, stacking, squeezing, and unsqueezing tensors
* Reshaping - reshapes an input tensor
* View - return a view of an imput tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other
* Squeeze - remove all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - return a view of the input with dimensions permuted (swapped) in a certain way


```python
x = torch.arange(1.,11.)
x,x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(2,5) # has to have same number of elements as input tensor!
x_reshaped,x_reshaped.size()
```

```python
# Change the view
z = x.view(1,10)
z,z.shape
```

```python
# Changing z changes x as a view of a tensor shares the same memory as the original input
z[:,0] = 5
z,x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked1 = torch.stack([x,x,x,x], dim=1)
x_stacked,x_stacked1
```

```python
# torch.squeeze() removes all dimensions with size `1`...think of this as "simplifying" a tensor down to the least dimensions needed
x=torch.rand(3,1,3)
y=torch.squeeze(x)
print(f"x = {x}\nx.shape = {x.shape}\n")
print(f"y = {y}\ny.shape = {y.shape}\n")
```

```python
# torch.unsqueeze() adds a single dimension to a tensor at a specific dimension
z=torch.unsqueeze(y,1)
print(f"y = {y}\ny.shape = {y.shape}\n")
print(f"z = {z}\nz.shape = {z.shape}")
```

```python
# torch.permute() rearranges the dimensions of a target tensor in a specified order
a=torch.rand((1,2,3))
b=torch.permute(a,(2,0,1))
print(f"{a}{a.shape}")
print(f"{b}{b.shape}")
```
